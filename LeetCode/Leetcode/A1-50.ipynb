```python

## LEETCODE 10
class Solution(object):
    def isMatch(self, s, p):
        # The DP table and the string s and p use the same indexes i and j, but
        # table[i][j] means the match status between p[:i] and s[:j], i.e.
        # table[0][0] means the match status of two empty strings, and
        # table[1][1] means the match status of p[0] and s[0]. Therefore, when
        # refering to the i-th and the j-th characters of p and s for updating
        # table[i][j], we use p[i - 1] and s[j - 1].

        # Initialize the table with False. The first row is satisfied.
        table = [[False] * (len(s) + 1) for _ in range(len(p) + 1)]

        # Update the corner case of matching two empty strings.
        table[0][0] = True

        # Update the corner case of when s is an empty string but p is not.
        # Since each '*' can eliminate the charter before it, the table is
        # vertically updated by the one before previous. [test_symbol_0]
        for i in range(2, len(p) + 1):
            table[i][0] = table[i - 2][0] and p[i - 1] == '*'

        for i in range(1, len(p) + 1):
            for j in range(1, len(s) + 1):
                if p[i - 1] != "*":
                    # Update the table by referring the diagonal element.
                    table[i][j] = table[i - 1][j - 1] and \
                                  (p[i - 1] == s[j - 1] or p[i - 1] == '.')
                else:
                    # Eliminations (referring to the vertical element)
                    # Either refer to the one before previous or the previous.
                    # I.e. * eliminate the previous or count the previous.
                    # [test_symbol_1]
                    table[i][j] = table[i - 2][j] or table[i - 1][j]

                    # Propagations (referring to the horizontal element)
                    # If p's previous one is equal to the current s, with
                    # helps of *, the status can be propagated from the left.
                    # [test_symbol_2]
                    if p[i - 2] == s[j - 1] or p[i - 2] == '.':
                        table[i][j] |= table[i][j - 1]

        return table[-1][-1]

solution = Solution()

s = "aa"
p = "a"
print(solution.isMatch(s,p))


s = "aa"
p = "a*"
print(solution.isMatch(s,p))


s = "ab"
p = ".*"
print(solution.isMatch(s,p))
```

```python
# Leetcode 31 
## https://leetcode.com/problems/next-permutation/discuss/14054/Python-solution-with-comments.
class Solution:
    def nextPermutation(self, nums):
        i = j = len(nums)-1
        while i > 0 and nums[i-1] >= nums[i]:
            i -= 1
        if i == 0:   # nums are in descending order
            nums.reverse()
            return 
        k = i - 1    # find the last "ascending" position
        while nums[j] <= nums[k]:
            j -= 1
        nums[k], nums[j] = nums[j], nums[k]  
        l, r = k+1, len(nums)-1  # reverse the second part
        while l < r:
            nums[l], nums[r] = nums[r], nums[l]
            l +=1 ; r -= 1
```

```python
# https://leetcode.com/problems/longest-valid-parentheses/discuss/2600732/Python3-Stack-DP-solution-with-thought-process.
# Leetcode 32
class Solution:
    def longestValidParentheses(self, s: str) -> int:       
        open_idxes = []  # Stack
        max_len_end_at = [0] * len(s)
        max_len_end_at.append(0) # To provide default value for -1 index access caused by parenthesis start with idx = 0. 
        
        for i, c in enumerate(s):
            if c == '(':
                open_idxes.append(i)
            else: # ")"
                if not open_idxes: # We can not create a valid subslice end at this index.
                    continue
                # Define current substring range.
                l = open_idxes.pop()
                r = i
                curr_len = r - l + 1
                  
                # Update DP array.
                max_len_end_at[i] = max_len_end_at[l-1] + curr_len
        
        return max(max_len_end_at) # Return global max.
```

```python
# Leetcode 34
class Solution:
    def searchRange(self, nums, target ):
        left = self.binSearch(nums, target, True)
        right = self.binSearch(nums, target, False)
        return [left,right]

    # leftBias = [T/F], if F, res is rightBiased
    def binSearch(self,nums, target,leftBias):
        l,r = 0, len(nums) - 1
        i = -1
        while l <= r:
            m = (l+r)//2
            if target  > nums[m]:
                l = m + 1
            elif target < nums[m]:
                r = m - 1
            else:
                i = m
                if leftBias:
                    r = m -1
                else:
                    l = m + 1
        return i

solution = Solution()
nums = [5,7,7,8,8,10]
target = 8
print(solution.searchRange(nums,target))


# nums = [5,7,7,8,8,10]
# target = 6
# print(solution.searchRange(nums,target))

# nums = []
# target = 0
# print(solution.searchRange(nums,target))
```

```python
# Leetcode 36
import collections

class Solution:
    def isValidSudoku(self, board ):
        cols = collections.defaultdict(set)
        rows = collections.defaultdict(set)
        squares = collections.defaultdict(set)

        for r in range(9):
            for c in range(9):
                if board[r][c] == ".":
                    continue
                if( board[r][c] in rows[r] or 
                    board[r][c] in cols[c] or
                    board[r][c] in squares[(r//3,c//3)]):
                    return False

                cols[c].add(board[r][c])
                rows[r].add(board[r][c])
                squares[(r//3,c//3)].add(board[r][c])

        return True
```

```python
# Leetcode 38
class Solution:
    def countAndSay(self, n: int) -> str:
        def countAndSayOneString(previous_string: str):
            current_string = ""

            current_digit = previous_string[0]
            count = 0
            for digit in previous_string:
                if digit != current_digit:
                    current_string += str(count) + current_digit
                    current_digit = digit
                    count = 0
                count += 1
            current_string += str(count) + current_digit

            return current_string

        answer = "1"
        for _ in range(n - 1):
            answer = countAndSayOneString(answer)

        return answer


solution = Solution()

n = 4
print(solution.countAndSay(n))

n = 1
print(solution.countAndSay(n))
```

```python
# Leetcode 39
class Solution:
    def combinationSum(self, candidates, target):
        res = []

        def dfs(i, cur, total):
            if total == target:
                res.append(cur.copy())
                return
            if i >= len(candidates) or total > target:
                return 

            cur.append(candidates[i])
            dfs(i, cur, total + candidates[i])
            cur.pop()
            dfs(i + 1, cur, total)

        dfs(0,[],0)
        return res

candidates = [2,3,6,7]
target = 7
solution = Solution()
solution.combinationSum(candidates, target)
            
```

```python
# leetcode 41 
# LINK https://leetcode.com/problems/first-missing-positive/discuss/2661122/Easy-to-understand-or-O(1)-Space-O(n)-Time-Python
class Solution:
    def firstMissingPositive(self, nums):
        def swap(i, j):
            nums[i], nums[j] = nums[j], nums[i]
            
        n = len(nums)
        for i in range(n):
            while nums[i] != i + 1:
                j = nums[i] - 1
                if 0 <= j < n and nums[j] != nums[i]:
                    swap(i, j)
                else:
                    break
        for i in range(n):
            if nums[i] != i + 1:
                return i + 1
        return n + 1
        
solution = Solution()
nums = [1,2,0]
print(solution.firstMissingPositive(nums))
nums = [3,4,-1,1]
print(solution.firstMissingPositive(nums))
nums = [7,8,9,11,12]
print(solution.firstMissingPositive(nums))
```

```python
# 43. Multiply Strings
class Solution:
    def multiply(self, num1: str, num2: str) -> str:
        offset = 1
        total = 0
        
        for digit1 in num1[::-1]:
            curr_total, curr, carry = 0, offset, 0
            for digit2 in num2[::-1]:
                new = int(digit1) * int(digit2) + carry
                carry, new = new // 10, new % 10

                curr_total += new * curr
                curr *= 10
            curr_total += curr * carry
            total += curr_total
            offset *= 10
        
        return str(total)
solution = Solution()
num1 = "7"
num2 = "5"
print(solution.multiply(num1,num2))

# num1 = "123"
# num2 = "456"
# print(solution.multiply(num1,num2))
```

```python
# 45. Jump Game II
class Solution:
    def jump(self, nums):
        if len(nums)==1:
            return 0
        l,r=1,nums[0]+1
        max_number=r-1
        count=1
        while max_number<len(nums)-1:
            for i in range(l,r):
                max_number=max(max_number,i+nums[i])
            l=r
            r=max_number+1
            count+=1
        return count

solution = Solution()

nums = [2,3,1,1,4]
print(solution.jump(nums))
nums = [2,3,0,1,4]
print(solution.jump(nums))
```

```python
# Leetcode 47
class Solution(object):
    def permuteUnique(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        result = []
        self.helper(  nums, [], len(nums), result)
        return result
    
    def helper(self, nums, so_far, N, result):
        if len(so_far) == N:
            result.append([x for x in so_far])
        else:
            candidates, cpos = set([]), {}
            for idx, x in enumerate(nums):
                if x not in candidates:
                    candidates.add(x)
                    cpos[x] = idx
       
            for c in candidates:
                so_far.append(c)
                nnums = nums[:cpos[c]] + nums[cpos[c]+1:]
                self.helper(nnums, so_far, N, result)
                so_far.pop()

        return
    
solution = Solution()
nums = [1,1,2]
print(solution.permuteUnique(nums))

# nums = [1,2,3]
# print(solution.permuteUnique(nums))
```

```python
## leetcode 49

from collections import defaultdict

class Solution(object):
    def groupAnagrams(self, strs):
        res = defaultdict(list)

        for s in strs:
            count = [0] * 26
            for c in s:
                count[ord(c) - ord("a")] += 1

            res[tuple(count)].append(s)
        return res.values()
```
