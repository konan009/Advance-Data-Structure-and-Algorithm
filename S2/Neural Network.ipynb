```python
import numpy as np
import random

def sigmoid(x):
    s = 1/(1+np.exp(-x))
    return s

def predict(dataset,weights):
    
    for data in dataset:
        row = data[:-1]
        row.append(1)

        signal = np.dot(row,weights)
        prediction = sigmoid(signal)

        if(prediction >= 0.5):
            print("DATA:",row[:-1],"Prediction:",1) 
        else:
            print("DATA:",row[:-1],"Prediction:",0) 


def train(train, l_rate, episodes,weights):
    for episode in range(episodes):
        sum_error = 0.0
        for data in train:
            #  GET THE DATA THEN SET FOR THE BIAS
            row = data[:-1]
            row.append(1)

            signal = np.dot(row,weights)
            prediction = sigmoid(signal)
            label = data[-1] 
            diff =  (prediction - label)**2
            sum_error += diff
            weights = np.array(weights) + (l_rate * (label - prediction) * np.array(row))

        mse = ( sum_error / len(dataset) )
        print('episode=%d,  mse=%.4f, ' % (episode, mse))

    return weights

# EXCERCISE 1 DATA
dataset = [ [0.3, 0],
            [0.1, 0],
            [0.5, 0],
            [0.6, 1],
            [0.7, 1],
            [0.8, 1]
            ]
weights = [0.5	,0.5]
l_rate = 0.8
episodes = 100
weights = train(dataset, l_rate, episodes,weights)
```

```python
## USE MODEL TO PREDICT
predict(dataset,weights)
```
