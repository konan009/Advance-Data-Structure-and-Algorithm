```python
"""
This block generate graph connection within each place where edge value is the time difference between each person
"""

generated_graph = {}

node_within = {}

# Loop the Places
for x in places_log:
    node_within = {}
    
    # LOOP THE LOGS WITHIN A PLACE
    for y in places_log[x]:
        user_id = y['user_id']
        timestamp = datetime.datetime.strptime(y['time'], "%Y-%m-%d %H:%M").timestamp()
        if user_id not in node_within:
            node_within[ user_id ] = { "time" : timestamp, "place" : x , "contact_no" : y['contact_info'] }

 
    for item in node_within:
        node_copy = node_within.copy()
        for node2 in node_copy:
          timestamp = node_copy[node2]["time"]

          # Check if the user has existing record
          if node2 in generated_graph:
            existing_connection = generated_graph[ node2 ]
          else:
            existing_connection = {}

          # If the current node within the array
          for node3 in node_copy:
            if(node3==node2):
              continue
            
            time_difference = timestamp - node_copy[node3]["time"]
            existing_connection[ abs(time_difference)  ] =   { "node" : node3, "place" : node_copy[node3]["place"], "contact_no" : node_copy[node3]['contact_no'] }
            
          generated_graph[ node2 ] = existing_connection

generated_graph
```

```python
""" 
Using networkx graph visualization for our connections of nodes
"""
import matplotlib.pyplot as plt
import networkx as nx

graph_with_labels = {}
graph_edges = []

""" 
O(NM) - Where N is users and M is the connection
"""

for item in generated_graph:
    for item2 in generated_graph[item]:
        graph_edges.append([item, generated_graph[item][item2]['node'] ])
        graph_with_labels[(item,generated_graph[item][item2]['node'])] = str( round(item2/3600,2) ) + "hr"
        
edges = graph_edges
edges_with_weights = graph_with_labels

G = nx.Graph()
G.add_edges_from(edges)
pos = nx.spring_layout(G)
plt.figure(figsize=(15, 15))
plt.title('Graph Generated from the Data')

nx.draw(
    G, pos, edge_color='black', width=1, linewidths=0.5,
    node_size=500, node_color='black', alpha=0.9,font_color='white',
    labels={node: node for node in G.nodes()}
)

nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edges_with_weights,
    font_color='blue')
plt.axis('off')
plt.show()
```

```python
"""
Perform Breadth-First Search
"""

INFECTED_USER_ID = 1
MAX_BREADTH = 2
HOUR_THRESHOLD = 3


visited = [INFECTED_USER_ID]
breadth_level = 1
flag = True
last_node = 13

queue = [INFECTED_USER_ID]

"""
O(NM) -  N is for the number of user and M edges of the node
"""
while len(queue) > 0:
    temp_queue = []
    print("Vulnerable Level : ",breadth_level)
    for current_node in queue:
        for edge in generated_graph[current_node]:
            node = generated_graph[current_node][edge]
            if((HOUR_THRESHOLD*3600)>edge and node['node'] not in visited):   
                print("  "," Person:", node['node']," place:",node['place']," contact no:",node['contact_no'])
                temp_queue.append(node['node'])
                visited.append(node['node'])
        
    breadth_level += 1
    if MAX_BREADTH < breadth_level:
        break
    
    queue = temp_queue
```
